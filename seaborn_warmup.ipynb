# Seaborn Warmup Exercise

````python
import seaborn as sns
from matplotlib import style

style.use("ggplot")

df = sns.load_dataset('mpg')
df.head()
````

Use seaborn to create the following visualizations:

1. Make a barplot of the average horsepower by origin. Add the keyword argument `ci=None`. What does this change?


````python
# make a barplot for average horsepower by origin
sns.barplot(x='origin', y='horsepower', data=df)
````

````python
sns.barplot(x='origin', y='horsepower', data=df, ci=None)
````

1. Make a scatterplot of displacement vs horsepower. Add the keyword argument `hue='weight'`. What does this change? Add the keyword argument `palette='Greens'`. What does this change?


````python
#make a scatterplot for displacment vs horsepower
sns.scatterplot(x='displacement', y='horsepower', data=df)
````

````python
sns.scatterplot(x='displacement', y='horsepower', data=df, hue = 'weight')
````

````python
sns.scatterplot(x='displacement', y='horsepower', data=df, hue = 'weight', palette='Greens')
````
